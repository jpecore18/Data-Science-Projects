# Ordered Songlist

I read in the smaller dataset of around 700 songs (a little less than that). All these songs are from 2020 and represent a wide variety of genres.

```python
ordered_songlist = pd.read_csv('ordered_songlist.csv')
```

```python
ordered_songlist.head()
```

# Ordered Large DF

I read in the larger dataset of around 150,000 songs (a little less than that). All these songs are from 2017-2020 and represent a wide variety of genres.

```python
#Data gathered partially from the dataset on https://www.kaggle.com/huanntran100/spotify-song-popularity-prediction
```

```python
ordered_large_df = pd.read_csv('ordered_large_df.csv')
```

# Concatenating Small DF and Large DF into Giant Ordered DF

I combine the two datasets of songs together so I can perform exploratory data analysis on them. This giant_ordered_df will later be used for both Regression and Classification modeling as well.

```python
giant_ordered_df = pd.concat([ordered_songlist, ordered_large_df])
```

```python
giant_ordered_df.head()
```

```python
giant_ordered_df.drop(columns=['key','index','genre','track_id'], inplace = True)
```

```python
giant_ordered_df.head()
```

# Exporting Giant Ordered DF

I export giant_ordered_df to csv. so I can use it to perform Regression and Classification modeling later.

```python
giant_ordered_df.to_csv('giant_ordered_df.csv')
```

# Exploratory Visual Data Analysis

I visually represent the data from all of the songs I can, exploring graphically how the giant_ordered_df is laid out in terms of popularity (by stream count), liveness, valence, acousticness, etc. I will later use these visualizations in my powerpoint presentation.

```python
song_features = ['danceability','energy','loudness','speechiness',
                 'instrumentalness', 'liveness', 'valence', 'tempo',
                'mode','duration_ms','time_signature','acousticness','time_signature']
```

```python
# Heatmap of all features correlation with "popularity"
corr_pop = giant_ordered_df.corr()[['popularity']]
plt.figure(figsize=(14,12))
sns.heatmap(corr_pop.sort_values('popularity',ascending=False), cmap="Greens", center=0, annot=True, vmin=-1, vmax=1);
```

```python
sns.distplot(giant_ordered_df['popularity'],color='blue').set_title('Popularity Distribution');
```

```python
giant_ordered_df.loc[giant_ordered_df['time_signature'] == 4] = 4/4
giant_ordered_df.loc[giant_ordered_df['time_signature'] == 3] = 3/4
giant_ordered_df.loc[giant_ordered_df['time_signature'] == 5] = 5/4
```

```python
giant_ordered_df['time_signature'].replace('0/4', 0, inplace=True)
giant_ordered_df['time_signature'].replace(0.25, 0, inplace=True)
giant_ordered_df['time_signature'].replace(1.25, 0, inplace=True)
giant_ordered_df['time_signature'].replace(1.0, 0, inplace=True)
giant_ordered_df['time_signature'].replace(0.75, 0, inplace=True)
```

```python
sns.barplot(x = 'time_signature', y = 'popularity', data = giant_ordered_df);
plt.title('Popularity Based on Time Signature')
```

```python
giant_ordered_df['mode'].replace(0.75, 0, inplace=True)
giant_ordered_df['mode'].replace(1.25, 1, inplace=True)
```

```python
sns.barplot(x = 'mode', y = 'popularity', data = giant_ordered_df);
plt.title('Popularity Based on Mode')
```

```python
sns.jointplot(x = 'acousticness', y = 'popularity', data = giant_ordered_df,color='orange');
```

```python
sns.jointplot(x = 'loudness', y = 'popularity', data = giant_ordered_df,color='red');
```

```python
sns.jointplot(x = 'liveness', y = 'popularity', data = giant_ordered_df,color='purple');
```

```python
sns.jointplot(x = 'energy', y = 'popularity', data = giant_ordered_df,color='blue');
```

```python
sns.jointplot(x = 'danceability', y = 'popularity', data = giant_ordered_df,color='black');
```

```python
sns.jointplot(x = 'tempo', y = 'popularity', data = giant_ordered_df,color='Green');
```

```python
popular_above_75 = giant_ordered_df[giant_ordered_df.popularity > 75]
sns.distplot(popular_above_75['acousticness'],color='orange')
plt.title('Acousticness for Songs with More than 75 Popularity');
```

```python
popular_below_75 = giant_ordered_df[giant_ordered_df.popularity < 75]
sns.distplot(popular_below_75['acousticness'],color='orange')
plt.title('Acousticness for Songs with Less than 60 Popularity');
```
